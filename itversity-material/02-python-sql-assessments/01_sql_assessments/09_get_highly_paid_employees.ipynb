```python
%%sql
SELECT department_name,employee_id,first_name,last_name,salary from (
SELECT d.department_name,e.employee_id,e.first_name,e.last_name,e.salary,
max(e.salary) OVER (PARTITION BY e.department_id ORDER BY e.salary DESC) as max_salary
FROM departments d RIGHT JOIN employees e
ON d.department_id = e.department_id
)as q
WHERE salary = max_salary
ORDER BY q.department_name,q.employee_id
```
